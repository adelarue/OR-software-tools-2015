##Aircraft Routing Problem to Minimize Propagated Delay##

### Flight Data  ###

We are going to solve a real-world aircraft routing problem for Alaska Airline!

"Flight_Alaska.csv", which contains an Alaska Airline daily flight schedule operated by Boeing 737 on January 1st, 2007.

There are 147 flights, 34 aircraft in total.

Each row represents a flight. From left to right, 

- Column 1: ID
- Column 2: Flight Number
- Column 3: Origin Airport Code
- Column 4: Destination Airport Code
- Column 5: Tail Number, a unique indicator for aircraft
- Column 6: Scheduled Departure Time, calculated as minutes from 01/01/1970
- Column 7: Scheduled Arrival Time, calculated as minutes from 01/01/1970
- Column 8: Flight Arrival Delay on 01/01/2007

```julia
flight_data=readcsv("Flight_Alaska.csv"; header=true)[1]
flight_header=readcsv("Flight_Alaska.csv"; header=true)[2]

println(flight_header)
println(flight_data)
```

###Construct Flight Connection Network###

Using this flight schedule, we can create the flight connection network we use to generate new columns (routes). To be specific, each node represents a flight. If flight A and flight B is connectable, i.e. flight B's departure airport is the same as flight A's arrival airport, flight B's departure time is later than flight A's arrival time plus minimum connection time, then a directed arc is established from flight node A to flight node B.

Don't forget to also add dummy source and sink node. Construct an edge from dummy source node to all flight nodes. Similarly, construct an edge from all flight nodes to dummy sink node.

```julia
using Graphs

#minimum connection time threshold
minimum_connection_time=30;

#construct flight graph 
flight_graph=graph(ExVertex[], ExEdge{ExVertex}[],is_directed=true)

#add vertices
for i=1:size(flight_data)[1]
   
    f_node=ExVertex(convert(Int,flight_data[i,1]), flight_data[i,2])
    f_node.attributes["origin"]=flight_data[i,3]
    f_node.attributes["destination"]=flight_data[i,4]
    f_node.attributes["tail_number"]=flight_data[i,5]
    f_node.attributes["sch_dep"]=flight_data[i,6]
    f_node.attributes["sch_arr"]=flight_data[i,7]
    f_node.attributes["delay"]=flight_data[i,8]
    
    #assign initial dual prices
    f_node.attributes["dual_prices"]=0.0
    add_vertex!(flight_graph, f_node)
    
end

#add flight connection edges
edge_index=1

for v1 in vertices(flight_graph)
    for v2 in vertices(flight_graph)
        if v1.attributes["sch_dep"]>=v2.attributes["sch_arr"]+minimum_connection_time && v1.attributes["origin"]==v2.attributes["destination"]
           
            conn_edge=ExEdge(edge_index, v1, v2)
            conn_edge.attributes["length"]=0.0
            add_edge!(flight_graph,conn_edge)
            edge_index=edge_index+1
            
        end
    end
end

#add dummy source and sink
source=ExVertex(num_vertices(flight_graph)+1, "source")
sink=ExVertex(num_vertices(flight_graph)+2, "sink")

add_vertex!(flight_graph, source)
add_vertex!(flight_graph, sink)

#Link dummy source to all flight nodes
for v in vertices(flight_graph)
    
    if v!=source && v!=sink
   
        source_edge=ExEdge(edge_index, source, v)
        source_edge.attributes["length"]=0.0
        add_edge!(flight_graph,source_edge)
        edge_index=edge_index+1
        
    end
    
end

#Link all flight nodes to dummy sink
for v in vertices(flight_graph)
   
    if v!=sink && v!=source
    
        sink_edge=ExEdge(edge_index, v, sink)
        sink_edge.attributes["length"]=0.0
        add_edge!(flight_graph,sink_edge)
        edge_index=edge_index+1
    
    end
    
end
```

Let's take a look at the flight connection network we just construct.

### Evaluate Alaska Airline's Original Route ###

Recall that in "Flight_Alaska.csv", there is a field called "Tail_Number", which is a unique indicator for the aircraft used to fly that flight. With that, we can first evaluate airline's original routes with respect to total propagated delay. We define a function "cal_propagate_delay" to calculate total propagated delay along one aircraft route.

```julia
function cal_propagate_delay(flight_ID_list)
    
    p_delay=0
    t_delay=0
    
    if length(flight_ID_list)==1
       
        return p_delay
        
    else
           
        for i=1:length(flight_ID_list)

            if i==1
               
                t_delay=vertices(flight_graph)[flight_ID_list[i]].attributes["delay"]
            
            else
                
                conn_time=vertices(flight_graph)[flight_ID_list[i]].attributes["sch_dep"]-vertices(flight_graph)[flight_ID_list[i-1]].attributes["sch_arr"]
                p_delay=max(0, -(conn_time-minimum_connection_time-t_delay))
                t_delay=p_delay+vertices(flight_graph)[flight_ID_list[i]].attributes["delay"]
                
            end

        end
        
        return p_delay
        
    end
    
end
```

Take a look at data set "Historical_Route.csv", where I reorder the rows by combining flights with same tailnumber together, and order them according to arrival time. Reading this data set, we can get the Alaska Airline's original routes for these 147 flights. And evaluate the total propagated delay over all routes using "cal_propagate_delay" function.

```julia
historical_route=readcsv("Historical_Route.csv"; header=true)[1]

previous_tailnumber=""

A=Float64[]
column=zeros(size(historical_route)[1])
column_ID=Int[]
c=Int[]

for i=1:size(historical_route)[1]
    
    if historical_route[i,5]!=previous_tailnumber
        if i!=1
            append!(A,column)
            push!(c, cal_propagate_delay(column_ID))
        end
        column=zeros(size(historical_route)[1])
        column_ID=Int[]
        column[convert(Int,historical_route[i,1])]=1
        push!(column_ID,convert(Int,historical_route[i,1]))
        previous_tailnumber=historical_route[i,5]
    
    else
        
        column[convert(Int,historical_route[i,1])]=1
        push!(column_ID,convert(Int,historical_route[i,1]))
        previous_tailnumber=historical_route[i,5]
        
    end
    
end

append!(A,column)
push!(c, cal_propagate_delay(column_ID))

A=reshape(A, size(historical_route)[1], convert(Int,length(A)/size(historical_route)[1]))
println("Total Propagated Delay over 147 Flights under Alaska Airline's Original Route: ", sum(c), " minutes")
```

### Construct the Restricted Master Problem ###

The total propagated delay over 147 flights under Alaska Airline's original route is 648. Let's see how we could reduce it.

Same as the cutting stock problem, we are going to solve the LP relaxation for the original aircraft routing problem first and then obtain the integer solution using branch and bound on root node. The formulation is given as follows,

$$\begin{align}
\nonumber
\min\qquad&\sum_{r\in R}p_rx_r \\ 
\textrm{s.t.}\qquad&A\mathbb{x}=A_1x_1+A_2x_2+\cdots+A_{|R|}x_{|R|}=1 \\
&\sum_{r\in R}x_r\le n \\
&0\le x_r\le 1,~\forall r\in R
\end{align}$$

where $n$ is the number of available aircraft, $R$ is the set of all feasible routes, $A$ is the flight-route matrix, $A_r$ is the $r$th column of $A$.

We know that $|R|$ is huge, column generation needs to be applied to generate variables on the fly. To initiate the solution process, we need to have a set of initial columns to make the first restricted master problem feasible. We can just use Alaska Airline's original routes as our initial columns. (_Note_: This is very common in solving column generation problem where you usually feed the first restricted master problem using current available solution. This could be current practice from the company, some naive solution you generate from a heuristic. Good initial solutions will greatly speed up the solution process.)

```julia
using JuMP, Gurobi

master = Model(solver=GurobiSolver(Method=0)) 

num_flights=147
num_aircraft=34

#define initial variables based on historical route
@defVar(master, 0<=x[1:length(c)]<=1)

#define constraint references, 147 flight cover constraints, 1 aircraft constraints
@defConstrRef myCons[1:(num_flights+1)]

#define constraints
#flight cover constraints
for i=1:num_flights
    myCons[i] = @addConstraint(master, dot(x, vec(A[i,:]))==1)
end

#aircraft constraints
myCons[num_flights+1] = @addConstraint(master, sum(x)<=num_aircraft)

#define objective
@setObjective(master, Min, dot(x,c))

solve(master)
```

####Updating Dual Prices on Flight Connection Network####

We are now ready to generate new columns based on the optimal dual prices. Recall the sub-problem here is a shortest path problem. So we need to update edge length based on the new dual prices we get. We define a function "update_network_dual" to accomplish that.

```julia
function update_network_dual(dual_prices)
    
    for i=1:num_flights
       
        vertices(flight_graph)[i].attributes["dual_prices"]=dual_prices[i]
        
    end
    
    # updating edge length based on new dual prices

    for v in vertices(flight_graph) 
       
        if v!=source && v!=sink
           
            for e in in_edges(v, flight_graph)
                
                e.attributes["length"]=-v.attributes["dual_prices"]
                
            end
            
        end
        
    end
    
    for e in in_edges(sink, flight_graph)

        e.attributes["length"]=-dual_prices[num_flights+1]

    end
    
end

#get optimal dual prices
r=[getDual(myCons)[1:num_flights+1]]

#update dual price on flight network
update_network_dual(r)
```

### Solving Shortest Path Sub-problem to Generate New Column (Route) ###

We then solve the shortest path problem from dummy origin to dummy sink, get the optimal s-t path. We then calculate the reduced cost of that optimal route by summing the optimal s-t distance and its total propagated delay.

```julia
edgel=AttributeEdgePropertyInspector{Float64}("length")

#Bellman Ford Algorithm to Calculate Shortest Path, why not Dijkstra?
path=bellman_ford_shortest_paths(flight_graph, edgel, [source])

iterator=path.parents[vertex_index(sink, flight_graph)]

column_ID=Int[]

while vertex_index(iterator, flight_graph)!=vertex_index(source, flight_graph)
    
    push!(column_ID, vertex_index(iterator, flight_graph))
    
    iterator=path.parents[vertex_index(iterator, flight_graph)]
    
end

#calculate the reduced cost by summing the optimal s-t distance and its total propagated delay.
reduced_cost=cal_propagate_delay(column_ID)+path.dists[vertex_index(sink, flight_graph)]

println("reduced cost of the new route is: ", reduced_cost)
```

```julia
# Construct the new column
column=zeros(num_flights)

for i=1:length(column_ID)
   
    column[column_ID[i]]=1
    
end

push!(column, 1)


# Add the new column
@defVar(master, 0<=z<=1, objective=cal_propagate_delay(column_ID), inconstraints=myCons, coefficients=column)
println("New Column(Route) Added with Flight Order: ")

for i=1:length(column_ID)
   
    print(vertices(flight_graph)[column_ID[i]].label)
    print(" origin: ", vertices(flight_graph)[column_ID[i]].attributes["origin"])
    print(" destination: ", vertices(flight_graph)[column_ID[i]].attributes["destination"], "\n")

end

println()

solve(master)
```

```julia
using JuMP, Gurobi, Graphs

function aircraft_routing(num_aircraft)

flight_data=readcsv("Flight_Alaska.csv"; header=true)[1]
flight_header=readcsv("Flight_Alaska.csv"; header=true)[2]

minimum_connection_time=30;
num_flights=147


#construct flight graph 
flight_graph=graph(ExVertex[], ExEdge{ExVertex}[],is_directed=true)

#add vertices
for i=1:size(flight_data)[1]
   
    f_node=ExVertex(convert(Int,flight_data[i,1]), flight_data[i,2])
    f_node.attributes["origin"]=flight_data[i,3]
    f_node.attributes["destination"]=flight_data[i,4]
    f_node.attributes["tail_number"]=flight_data[i,5]
    f_node.attributes["sch_dep"]=flight_data[i,6]
    f_node.attributes["sch_arr"]=flight_data[i,7]
    f_node.attributes["delay"]=flight_data[i,8]
    
    #assign initial dual prices
    f_node.attributes["dual_prices"]=0.0
    add_vertex!(flight_graph, f_node)
    
end

#add flight connection edges
edge_index=1

for v1 in vertices(flight_graph)
    for v2 in vertices(flight_graph)
        if v1.attributes["sch_dep"]>=v2.attributes["sch_arr"]+minimum_connection_time && v1.attributes["origin"]==v2.attributes["destination"]
           
            conn_edge=ExEdge(edge_index, v1, v2)
            conn_edge.attributes["length"]=0.0
            add_edge!(flight_graph,conn_edge)
            edge_index=edge_index+1
            
        end
    end
end

#add dummy source and sink
source=ExVertex(num_vertices(flight_graph)+1, "source")
sink=ExVertex(num_vertices(flight_graph)+2, "sink")

add_vertex!(flight_graph, source)
add_vertex!(flight_graph, sink)

#Link dummy source to the other nodes
for v in vertices(flight_graph)
    
    if v!=source && v!=sink
   
        source_edge=ExEdge(edge_index, source, v)
        source_edge.attributes["length"]=0.0
        add_edge!(flight_graph,source_edge)
        edge_index=edge_index+1

    end
    
end

#Link dummy sink to the other nodes
for v in vertices(flight_graph)
   
    if v!=sink && v!=source
    
        sink_edge=ExEdge(edge_index, v, sink)
        sink_edge.attributes["length"]=0.0
        add_edge!(flight_graph,sink_edge)
        edge_index=edge_index+1
    
    end
    
end

function cal_propagate_delay(flight_ID_list)
    
    p_delay=0
    t_delay=0
    
    if length(flight_ID_list)==1
       
        return p_delay
        
    else
           
        for i=1:length(flight_ID_list)

            if i==1
               
                t_delay=vertices(flight_graph)[flight_ID_list[i]].attributes["delay"]
            
            else
                
                conn_time=vertices(flight_graph)[flight_ID_list[i]].attributes["sch_dep"]-vertices(flight_graph)[flight_ID_list[i-1]].attributes["sch_arr"]
                p_delay=max(0, -(conn_time-minimum_connection_time-t_delay))
                t_delay=p_delay+vertices(flight_graph)[flight_ID_list[i]].attributes["delay"]
                
            end

        end
        
        return p_delay
        
    end
    
end


historical_route=readcsv("Historical_Route.csv"; header=true)[1]

initial_columns_str=String[]
new_columns_str=String[]

previous_tailnumber=""


A=Float64[]
column=zeros(size(historical_route)[1])
column_ID=Int[]
c=Int[]

for i=1:size(historical_route)[1]
    
    if historical_route[i,5]!=previous_tailnumber
        if i!=1
            append!(A,column)
            push!(c, cal_propagate_delay(column_ID))
            column_str="Routing One Aircraft Through: "
            for j=1:length(column_ID)
                column_str=string(column_str, vertices(flight_graph)[column_ID[j]].label, ",", vertices(flight_graph)[column_ID[j]].attributes["origin"], ",", vertices(flight_graph)[column_ID[j]].attributes["destination"],";")
            end
            push!(initial_columns_str, column_str)
        end
        column=zeros(size(historical_route)[1])
        column_ID=Int[]
        column[convert(Int,historical_route[i,1])]=1
        push!(column_ID,convert(Int,historical_route[i,1]))
        previous_tailnumber=historical_route[i,5]
    
    else
        
        column[convert(Int,historical_route[i,1])]=1
        push!(column_ID,convert(Int,historical_route[i,1]))
        previous_tailnumber=historical_route[i,5]
        
    end
    
end

append!(A,column)
push!(c, cal_propagate_delay(column_ID))

column_str="Routing One Aircraft Through: "
for j=1:length(column_ID)
    column_str=string(column_str, vertices(flight_graph)[column_ID[j]].label, ",", vertices(flight_graph)[column_ID[j]].attributes["origin"], ",", vertices(flight_graph)[column_ID[j]].attributes["destination"],";")
end
push!(initial_columns_str, column_str)

A=reshape(A, size(historical_route)[1], convert(Int,length(A)/size(historical_route)[1]))



master = Model(solver=GurobiSolver(Method=0, OutputFlag=0)) 


#define initial variables based on historical route
@defVar(master, 0<=x[1:length(c)]<=1)

#define constraint references
@defConstrRef myCons[1:(num_flights+1)]

#define constraints
for i=1:num_flights
    myCons[i] = @addConstraint(master, dot(x, vec(A[i,:]))==1)
end

myCons[num_flights+1] = @addConstraint(master, sum(x)<=num_aircraft)

#define objective
@setObjective(master, Min, dot(x,c))

solve(master)

#get optimal dual prices
r=[getDual(myCons)[1:num_flights+1]]

#update dual price on flight network
function update_network_dual(dual_prices)
    
    for i=1:num_flights
       
        vertices(flight_graph)[i].attributes["dual_prices"]=dual_prices[i]
        
    end
    
    for v in vertices(flight_graph) 
       
        if v!=source && v!=sink
           
            for e in in_edges(v, flight_graph)
                
                e.attributes["length"]=-v.attributes["dual_prices"]
                
            end
            
        end
        
    end

    for e in in_edges(sink, flight_graph)

        e.attributes["length"]=-dual_prices[num_flights+1]

    end
    
end

update_network_dual(r)

edgel=AttributeEdgePropertyInspector{Float64}("length")
path=bellman_ford_shortest_paths(flight_graph, edgel, [source])

iterator=path.parents[vertex_index(sink, flight_graph)]

column_ID=Int[]

while vertex_index(iterator, flight_graph)!=vertex_index(source, flight_graph)
    
    push!(column_ID, vertex_index(iterator, flight_graph))
    
    iterator=path.parents[vertex_index(iterator, flight_graph)]
    
end

reduced_cost=cal_propagate_delay(column_ID)+path.dists[vertex_index(sink, flight_graph)]

epsilon=1e-6

newColumns=Variable[]

while reduced_cost<=-epsilon
    
    column=zeros(num_flights)

    for i=1:length(column_ID)

        column[column_ID[i]]=1

    end

    push!(column, 1)

    @defVar(master, 0<=z<=1, objective=cal_propagate_delay(column_ID), inconstraints=myCons, coefficients=column)
    
    println("New Column(Route) Added with Reduced Cost: ", reduced_cost, "\nFlight Order: ")

    column_str_2="Routing One Aircraft Through: "
    for i=1:length(column_ID)
        column_str_2=string(column_str_2, vertices(flight_graph)[column_ID[i]].label, ",", vertices(flight_graph)[column_ID[i]].attributes["origin"], ",", vertices(flight_graph)[column_ID[i]].attributes["destination"],";")
    end
    push!(new_columns_str, column_str_2)

    for i=1:length(column_ID)

        print(vertices(flight_graph)[column_ID[i]].label)
        print(" origin: ", vertices(flight_graph)[column_ID[i]].attributes["origin"])
        print(" destination: ", vertices(flight_graph)[column_ID[i]].attributes["destination"], "\n")
        
    end

    println()

    push!(newColumns, z)

    solve(master)
    
    #get optimal dual prices
    r=[getDual(myCons)[1:num_flights+1]]
    
    update_network_dual(r)

    edgel=AttributeEdgePropertyInspector{Float64}("length")
    path=bellman_ford_shortest_paths(flight_graph, edgel, [source])

    iterator=path.parents[vertex_index(sink, flight_graph)]

    column_ID=Int[]

    while vertex_index(iterator, flight_graph)!=vertex_index(source, flight_graph)

        push!(column_ID, vertex_index(iterator, flight_graph))

        iterator=path.parents[vertex_index(iterator, flight_graph)]

    end



    reduced_cost=cal_propagate_delay(column_ID)+path.dists[vertex_index(sink, flight_graph)]
    
end


println("Optimal Objective Value of LP Relaxation:", getObjectiveValue(master))
println("Total Number of Columns:", length(x)+length(newColumns))

#change variable type from continuous to integer

for i=1:length(x)
    setCategory(x[i], :Bin)
end

for i=1:length(newColumns)
    setCategory(newColumns[i],:Bin)
end

solve(master)

println("Optimal Objective Value of IP Based on Root Node Branch and Bound :", getObjectiveValue(master))

println("\nOptimal Routes are:\n")

for i=1:length(x)
    if getValue(x[i])>epsilon
        println(initial_columns_str[i])
    end
end

for i=1:length(newColumns)
    if getValue(newColumns[i])>epsilon
        println(new_columns_str[i])
    end
end
    
end
```
