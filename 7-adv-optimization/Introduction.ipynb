# Mixed-integer programming: Tools and tricks

Operations research often involves models where you need discrete decisions. 

Oftentimes, this will lead to problems that are $NP$-hard. What do you do?

1. Give up
2. Settle for approximate solutions
2. Try the simplest thing imaginable: **enumeration**

Consider the simple knapsack. Iain wants to carry items to the pawn shop to get some extra cash. He has $N$ items, each with a weight $w_i$ and a price $p_i$. Iain hasn't been to the gym lately, so he can only carry $C$ kilos. How does he choose what to bring with him?

We can model this as an integer optimization problem:

\begin{align*}
\max& \sum_{i=1}^N p_i x_i \\
\text{s.t.}& \sum_{i=1}^N w_i x_i \leq C \\
& 0 \leq x_i \leq 1 \quad \forall i = 1,\ldots,N \\
& x_i \in \{0,1\} \quad \forall i = 1,\ldots,N
\end{align*}

How would you solve this? The simple way is just to consider each possible value for $x$ and compare the cost. After Iain has weighed all $2^N$ possible collections of items (and verified that he can lift them at once), he just chooses the best set.

One way of formalizing this counting scheme is to consider a partition of $\{1,\ldots,N\}$ given by $I^0$ and $I^1$. We can bundle together the act of assessing the payoff and making sure Iain can carry the pack by solving the trivial linear program

\begin{align*}
\max& \sum_{i=1}^N p_ix_i \\
\text{s.t.}& \sum_{i=1}^N w_i x_i \leq C \\
& 0 \leq x_i \leq 1 \quad \forall i = 1,\ldots,N \\
& x_i = 0 \quad \forall i \in I^0 \\
& x_i = 1 \quad \forall i \in I^1
\end{align*}

So if we just solve the LPs for each possible choice of $(I^0,I^1)$ and pick the partition with the greatest optimal cost, we're done!

We can visualize this approach to enumeration:
![Enumerate](http://www.mit.edu/~huchette/.tmp/enumerate.svg)

This might seem like we're just throwing extra mathematical complexity at a simple counting problem, but now we have the framework for looking at disjoint sets $I^0,I^1$ that don't necessarily partition $\{1,\ldots,N\}$. Maybe we say that we consider sets $I^0,I^1$ such that the two combined have exactly $k$ indices in them. Now our subproblem isn't so trivial; we actually have to _solve_ it. It also doesn't necessarily produce something that's integer feasible: the optimal solution might have some value $0 < x_i < 1$ for $i \notin I^0 \cup I^1$. Nevertheless, we still get some useful information. 

What if we construct a binary tree by iteratively choosing a variable and create two branches: one for each possible integral value it takes? We'll talk about *node* ``q`` and its corresponding subsets $I_q^0,I_q^1$. The tree would probably look something like this:
![Full Tree](http://www.mit.edu/~huchette/.tmp/fulltree.svg)

This is even more mathematical complexity, and what does it buy us? Well, with the tree structure in place, we can make a simple observation about bounding. Since as we go down the arcs of the tree we restrict our problem more and more, we must have that:

>If node ``q`` is descended from node ``p``, we must have that the optimal cost of subproblem ``q`` is no more than that for node ``p``

This leads us to a powerful tool in solving these enumeration problems: 

>If I can show you that the optimal cost for subproblem ``q`` is _less_ than the optimal cost for the original problem, the same is true for any descendent of ``q``. 

That is, we can _prune_ the tree and safely discard some nodes, kind of like this:
![Pruning](http://www.mit.edu/~huchette/.tmp/fathom.svg)

## Branch-and-bound
We'll keep track of a _lower bound_ $LB$ for our problem. Each node ``q`` will have an upper bound $UB_q$ that it inherents from its parents optimal cost. If we get to the point where we have solved all subproblems (or, ideally, pruned off a great deal of them), we know that we're optimal. To do this we'll also keep track of a list $L$ of subproblems left to solve; at the beginning, we'll take it to be the set of all possible $I^0,I^1$ pairs of any size. The procedure is:

1. Pick a subproblem ``q`` out of our list $L$ and solve it.
2. If the problem is infeasible, ``goto`` 1.
3. If the solution is integer feasible, update the lower bound $LB$ if the cost is higher than what we had before and ``goto 1``. 
4. If the relaxation value is less than our global $LB$, ``goto`` 1.
5. Otherwise, pick a non-integer variable $i$ and _branch_ by adding two subproblems to $L$: 
    * One by adding $i$ to $I_q^0$
    * Another by adding $i$ to $I_q^1$
      
      The upper bound value for each new node will be the objective value for node ``q``.
6. Repeat until $L$ is empty.

Branch-and-bound is sometimes called an _implicit enumeration_ scheme because of step 4: we avoid solving any subproblems that we can prove won't produce the optimal solution.

Consider the following (small) knapsack problem:

\begin{align*}
    \max\:& x + y \\
    \text{s.t.}\:& x + 2y \leq 1.5 \\
    & 0 \leq x,y \leq 1 \\
    & x, y \in \{0,1\}
\end{align*}

```julia
using Gadfly, Compose
xx = -1:0.01:2
yy = 0.5*(1.5-xx)
plot(
    layer(x=xx, y=yy, Geom.line),
    layer(xintercept=[0], Geom.vline()),
    layer(xintercept=[1], Geom.vline()),
    layer(yintercept=[0], Geom.hline()),
    layer(yintercept=[1], Geom.hline()),
    Guide.annotation(
        compose(context(), polygon([(0,0),(1,0),(1,.25),(0,.75)]),fill("bisque"))),
    Coord.cartesian(fixed=true)
)
```

First, we solve the LP relaxation and get $(x^*,y^*) = (1,0.25)$. 

```julia
plot(
    layer(x=xx, y=yy, Geom.line),
    layer(xintercept=[0], Geom.vline()),
    layer(xintercept=[1], Geom.vline()),
    layer(yintercept=[0], Geom.hline()),
    layer(yintercept=[1], Geom.hline()),
    Guide.annotation(
        compose(context(), circle([1], [.25], [1mm]), fill("red"),
               stroke("orange"))),
    Guide.annotation(
        compose(context(), polygon([(0,0),(1,0),(1,.25),(0,.75)]),fill("bisque"))),
    Coord.cartesian(fixed=true)
)
```

This isn't integer feasible, so we branch on $y$. The subproblem with $y = 1$ is infeasible, and the subproblem with $y = 0$ is feasible with solution $(x^*,y^*) = (1,0)$. This is integer feasible, so we update our lower bound. We've also exhausted the tree, so we have our optimal solution!

```julia
δ = 0.05
hstack(
    plot(
        layer(x=xx, y=yy, Geom.line),
        layer(xintercept=[0], Geom.vline()),
        layer(xintercept=[1], Geom.vline()),
        layer(yintercept=[0], Geom.hline()),
        layer(yintercept=[1], Geom.hline()),
        Guide.annotation(
            compose(context(), circle([1], [0], [1mm]), fill("red"),
                   stroke("orange"))),
        Guide.annotation(
compose(context(), polygon([(0-δ,0-δ),(1+δ,0-δ),(1+δ,0+δ),(0-δ,0+δ)]),fill("bisque"))),
        Coord.cartesian(fixed=true),
        Guide.title("y = 0")
    ),
    plot(
        layer(x=xx, y=yy, Geom.line),
        layer(xintercept=[0], Geom.vline()),
        layer(xintercept=[1], Geom.vline()),
        layer(yintercept=[0], Geom.hline()),
        layer(yintercept=[1], Geom.hline()),
        Coord.cartesian(fixed=true),
        Guide.title("y = 1")
    )
)
```

In the worst case scenario, the branch-and-bound scheme can end up solving even _more_ subproblems than if we just enumerated everything to begin with. For the scheme to work well, we need to prune large portions of the tree. 

** The "magic" of modern MIP solvers largely comes down to pruning massive portions of the tree **

We'll discuss four ways that we may prune our tree:

1. Cuts (improve the _lower bounds_)
2. Heuristics (improve the _upper bounds_)
3. Presolve (_simplify_ everything before we build the tree)
4. Branching strategies (construct the tree in a smart way)

## Cuts

Cuts are inequalities that are _valid_ (don't cut off any feasible points) and _strengthen_ the formulation (chop off some infeasible points that are included in our relaxation)

Three main types of cuts:

1. General purpose (e.g. CG, split, MIR)
2. Structure-specific (e.g. knapsack cover, clique)
3. Problem-specific (whatever _you_ cook up)

Let's go back to our silly knapsack example and imagine that had been a bit smarter and realized that $x + \frac{4}{3}y \leq 1$ is feasible for all integer feasible points. If we add this to our formulation, we have the optimal solution at the root node! Cuts are awesome!

```julia
ycut = 3/4*(1-xx)
plot(
    layer(x=xx, y=yy, Geom.line),
    layer(x=xx, y=ycut, Geom.line),
    layer(xintercept=[0], Geom.vline()),
    layer(xintercept=[1], Geom.vline()),
    layer(yintercept=[0], Geom.hline()),
    layer(yintercept=[1], Geom.hline()),
    Guide.annotation(
        compose(context(), polygon([(0,0),(1,0),(1,0),(0,.75)]),fill("red"))),
    Guide.annotation(
        compose(context(), polygon([(0,0),(1,0),(1,.25),(0,.75)]),fill("bisque"))),
    Coord.cartesian(fixed=true)
)
```

## Heuristics

You can inject feasible solutions into the branch-and-bound scheme to improve the upper bound. These can come from

1. Problem-specific heuristics
2. Neighborhood search
3. Rounding or "polishing"

After we've already solved it twice, it's easy to see that the optimal solution for our knapsack is $(1,0)$. If we inject this into our branch-and-bound scheme, we get that the root node optimal value value, at 1.5, is _greater_ than our lower bound, so we can just go ahead and prune the entire tree!

```julia
plot(
    layer(x=xx, y=yy, Geom.line),
    layer(xintercept=[0], Geom.vline()),
    layer(xintercept=[1], Geom.vline()),
    layer(yintercept=[0], Geom.hline()),
    layer(yintercept=[1], Geom.hline()),
    Guide.annotation(
        compose(context(), circle([1], [0], [1mm]), fill("red"),
               stroke("orange"))),
    Guide.annotation(
        compose(context(), polygon([(0,0),(1,0),(1,.25),(0,.75)]),fill("bisque"))),
    Coord.cartesian(fixed=true)
)
```

## Presolve

Modern MIP solvers have a litany of techniques to simplify problems _before_ constructing the tree. 

* Variable/constraint bounds tightening (more on this later!)
* Logical inferences (again, more later!)
* General cleanup (remove redundant variables, constraints)

By shrinking problems, presolve potentially shrinks

* the B&B tree (fewer binary variables)
* The node solve time (smaller LP relaxations)

First, we can perform bounds tightening to get

\begin{align*}
    \max\:& x + y \\
    \text{s.t.}\:& x + 2y \leq 1 \\
    & 0 \leq x,y \leq 1 \\
    & x, y \in \{0,1\}
\end{align*}

```julia
ytight = 0.5*(1-xx)
plot(
layer(x=xx, y=ytight, Geom.line),
    layer(xintercept=[0], Geom.vline()),
    layer(xintercept=[1], Geom.vline()),
    layer(yintercept=[0], Geom.hline()),
    layer(yintercept=[1], Geom.hline()),
    Guide.annotation(
        compose(context(), polygon([(0,0),(1,0),(1,0),(0,.5)]),fill("bisque"))),
    Coord.cartesian(fixed=true)
)
```

Then, we can reason that, since $x,y \in \{0,1\}$, we cannot have $y=1$, and so we can reduce our problem to 

\begin{align*}
    \max\:& x \\
    & x \in \{0,1\}
\end{align*}

Which we can then solve in closed form. Iain will show us an even more radical example later.

## Branching strategies

Different paths through the B&B tree can see drastically different performance:

* Missing a node that produces a good feasible solution will mean you add nodes to your list you otherwise would have pruned
* Choosing nodes that would otherwise be pruned just slows you down

How do we know _a priori_ what a good branching strategy is, though? 

We don't really, but you can make informed guesses. 

Consider another small problem:
\begin{align*}
    \max\:& y \\
    \text{s.t.}\:& -x + y \leq \frac{1}{3} \\
    & x + y \leq \frac{4}{3} \\
    & 0 \leq x,y \leq 1 \\
    & x, y \in \{0,1\}
\end{align*}
If we solve the relaxation, we get an optimal solution of $(x^*,y^*) = (\frac{1}{2},\frac{5}{6})$.

```julia
plot(
    layer(x=xx, y=1/3+xx, Geom.line),
    layer(x=xx, y=4/3-xx, Geom.line),
    layer(xintercept=[0], Geom.vline()),
    layer(xintercept=[1], Geom.vline()),
    layer(yintercept=[0], Geom.hline()),
    layer(yintercept=[1], Geom.hline()),
    Guide.annotation(
        compose(context(), circle([0.5], [5/6], [1mm]), fill("red"),
               stroke("orange"))),
    Guide.annotation(
        compose(context(), polygon([(0,0),(1,0),(1,1/3),(1/2,5/6),(0,1/3)]),fill("bisque"))),
    Coord.cartesian(fixed=true)
)
```

Now we have a choice: both coordinates are fractional, so we can branch on either. If we branch on $y$, we get

```julia
δ = 0.05
hstack(
    plot(
        layer(x=xx, y=1/3+xx, Geom.line),
        layer(x=xx, y=4/3-xx, Geom.line),
        layer(xintercept=[0], Geom.vline()),
        layer(xintercept=[1], Geom.vline()),
        layer(yintercept=[0], Geom.hline()),
        layer(yintercept=[1], Geom.hline()),
        Guide.annotation(
            compose(context(), circle([1], [0], [1mm]), fill("red"),
                   stroke("orange"))),
        Guide.annotation(
compose(context(), polygon([(0-δ,0-δ),(1+δ,0-δ),(1+δ,0+δ),(0+δ,0+δ)]),fill("bisque"))),
        Coord.cartesian(fixed=true),
        Guide.title("y = 0")
    ),
    plot(
        layer(x=xx, y=1/3+xx, Geom.line),
        layer(x=xx, y=4/3-xx, Geom.line),
        layer(xintercept=[0], Geom.vline()),
        layer(xintercept=[1], Geom.vline()),
        layer(yintercept=[0], Geom.hline()),
        layer(yintercept=[1], Geom.hline()),
        Coord.cartesian(fixed=true),
        Guide.title("y = 1")
    )
)
```

The branch with $y=0$ gives us an optimal solution in $(0,1)$, and the $y=1$ branch is infeasible, so we have no more branches to take and we are done. But if we had branched on $x$ instead,

```julia
δ = 0.05
hstack(
    plot(
        layer(x=xx, y=1/3+xx, Geom.line),
        layer(x=xx, y=4/3-xx, Geom.line),
        layer(xintercept=[0], Geom.vline()),
        layer(xintercept=[1], Geom.vline()),
        layer(yintercept=[0], Geom.hline()),
        layer(yintercept=[1], Geom.hline()),
        Guide.annotation(
            compose(context(), circle([0], [1/3], [1mm]), fill("red"),
                   stroke("orange"))),
        Guide.annotation(
compose(context(), polygon([(0-δ,0-δ),(0+δ,0-δ),(0+δ,1/3+δ),(0-δ,1/3+δ)]),fill("bisque"))),
        Coord.cartesian(fixed=true),
        Guide.title("x = 0")
    ),
    plot(
        layer(x=xx, y=1/3+xx, Geom.line),
        layer(x=xx, y=4/3-xx, Geom.line),
        layer(xintercept=[0], Geom.vline()),
        layer(xintercept=[1], Geom.vline()),
        layer(yintercept=[0], Geom.hline()),
        layer(yintercept=[1], Geom.hline()),
        Guide.annotation(
            compose(context(), circle([1], [1/3], [1mm]), fill("red"),
                   stroke("orange"))),
        Guide.annotation(
compose(context(), polygon([(1-δ,0-δ),(1+δ,0-δ),(1+δ,1/3+δ),(1-δ,1/3+δ)]),fill("bisque"))),
        Coord.cartesian(fixed=true),
        Guide.title("x = 1")
    )
)
```

The optimal solution for both branches is still fractional, so we have to branch again. If we have, say, 1000 binary variables instead of just two, we see how branching strategies can have an enormous effect on performance.
